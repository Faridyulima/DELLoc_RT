```python
import matplotlib.pyplot as plt
import numpy as np

# Data
datasets = ['UJIIndoorLoc', 'UTSIndoorLoc', 'Tampere']
with_confidence = [93.32, 94.38, 94.02]
without_confidence = [91.61, 93.56, 93.44]

x = np.arange(len(datasets))
width = 0.3

# Create figure
plt.figure(figsize=(5, 5))

# Bars
bars1 = plt.bar(x - width/2, with_confidence, width,
                label='with Confidence Weighting',
                color='skyblue', edgecolor='black')

bars2 = plt.bar(x + width/2, without_confidence, width,
                label='without Confidence Weighting',
                color='lightcoral', edgecolor='black')

# Title and labels
plt.title('Floor Identification')
plt.ylabel('Accuracy (%)')
plt.xticks(x, datasets)
plt.ylim(91, 96)
plt.legend()

# Add value labels on bars
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2,
                 height + 0.05,
                 f'{height:.2f}',
                 ha='center', va='bottom', fontsize=9)

# Layout and show
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel('CDF_Tampere_ML.xlsx')

knn = df.iloc[:3514,0:1]
rf = df.iloc[:3514,1:2]
lgb = df.iloc[:3514,2:3]
xgb = df.iloc[:3514,3:4]
ada = df.iloc[:3514,4:5]
mlp = df.iloc[:3514,5:6]
delloc = df.iloc[:3514,6:7]

# getting data of the histogram
count1, bins_count1 = np.histogram(knn, bins=300)
count2, bins_count2 = np.histogram(rf, bins=300)
count3, bins_count3 = np.histogram(lgb, bins=300)
count4, bins_count4 = np.histogram(xgb, bins=300)
count5, bins_count5 = np.histogram(ada, bins=300)
count6, bins_count6 = np.histogram(mlp, bins=300)
count7, bins_count7 = np.histogram(delloc, bins=300)

pdf1 = count1 / sum(count1)
pdf2 = count2 / sum(count2)
pdf3 = count3 / sum(count3)
pdf4 = count4 / sum(count4)
pdf5 = count5 / sum(count5)
pdf6 = count6 / sum(count6)
pdf7 = count7 / sum(count7)

cdf1 = np.cumsum(pdf1)
cdf2 = np.cumsum(pdf2)
cdf3 = np.cumsum(pdf3)
cdf4 = np.cumsum(pdf4)
cdf5 = np.cumsum(pdf5)
cdf6 = np.cumsum(pdf6)
cdf7 = np.cumsum(pdf7)

plt.figure(figsize=(7, 6))  # Adjust width and height as needed

plt.plot(bins_count1[1:], cdf1, label="KNN [46]", linewidth=1.5, color = '#a2142f')
plt.plot(bins_count2[1:], cdf2, label="Random Forest [47]", linewidth=1.5, color = '#fc0406')
plt.plot(bins_count3[1:], cdf3, label="LightGBM [48]", linewidth=1.5, color = '#f26f2a')
plt.plot(bins_count4[1:], cdf4, label="XGBoost [49]", linewidth=1.5, color = '#77ac30')
plt.plot(bins_count5[1:], cdf5, label="AdaBoost [50]", linewidth=1.5, color = '#00bfbf')
plt.plot(bins_count6[1:], cdf6, label="MLP [23]", linewidth=1.5, color = '#0072bd')
plt.plot(bins_count7[1:], cdf7, label="DELLoc", linewidth=1.5, color = '#282e81')

plt.ylim([0, 1])
plt.xlim([0, 50])
plt.grid(linestyle = '--', linewidth = 0.5)
plt.title('Tampere', fontsize=16)
plt.xlabel('Euclidean Error (m)', fontsize = 14)
plt.ylabel('CDF', fontsize = 14)
plt.xticks(fontsize=12)                         
plt.yticks(fontsize=12)
plt.legend(fontsize=14, loc = 4)
plt.show()
```
